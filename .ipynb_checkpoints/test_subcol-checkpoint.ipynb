```python
%%cython
import numpy, warnings
def calc_overlap(q1, q2, xaxis=0, cf_min=0.0, cf_max=1.0):            
    
    # cloud fractions for each layer
    c1 = (q1 > 0).mean(axis=xaxis)
    c2 = (q2 > 0).mean(axis=xaxis)
            
    # theoretical combined cloud fractions
    cmax = numpy.maximum(c1, c2)
    cran = c1 + c2 - (c1 * c2)
    ctrue = ((q1 > 0) | (q2 > 0)).mean(axis=xaxis)
    
    # overlap parameter alpha
    alpha = numpy.where(cmax != cran, (ctrue - cran)/(cmax - cran), 
                        numpy.nan)
    
    # filter out cases with small or large cloud fractions?
    # Tompkins and Di Giuseppe suggest only considering scenes for which
    # cloud scale is smaller than sampling scale...so only partly-cloudy
    # layers?
    alpha = numpy.where((c1 > cf_min) & (c2 > cf_min) & 
                        (c1 < cf_max) & (c2 < cf_max), 
                        alpha, numpy.nan)
    
    return alpha


def calc_rankcorr(q1, q2, xaxis=0):
    # calculate ranks for OVERLAPPING CLOUDY PARTS of both layers
    overlapping = (q1 > 0) & (q2 > 0)
    r1 = numpy.where(overlapping, q1, numpy.nan).argsort(axis=xaxis).argsort(axis=xaxis)
    r2 = numpy.where(overlapping, q2, numpy.nan).argsort(axis=xaxis).argsort(axis=xaxis)

    # want only the valid ranks, not the ranks where there is no cloud!
    # including those ranks would artificially inflate rho...
    # NOTE: MUST INCLUDE THESE LINES! The sort method above gives ranks
    # to ALL points, even those defined as NaN, the NaNs are just moved
    # to the end. So, unless we mask these out again AFTER sorting, we
    # would still be artificially inflating rho, even though we masked
    # q1 and q2 already.
    x = numpy.where(overlapping, r1, numpy.nan)
    y = numpy.where(overlapping, r2, numpy.nan)

    # calculate pearson product-moment correlation coefficient of ranks
    with warnings.catch_warnings():
        warnings.simplefilter('ignore', category=RuntimeWarning)
        xb = numpy.nanmean(x, axis=xaxis, keepdims=True)
        yb = numpy.nanmean(y, axis=xaxis, keepdims=True)
        rho = numpy.nansum((x - xb)*(y - yb), axis=xaxis)/numpy.sqrt(
            numpy.nansum((x - xb)**2, axis=xaxis) * numpy.nansum((y - yb)**2, axis=xaxis)
        )
    
        # filter out cases with small cloud fraction?
        rho = numpy.where(numpy.nanmean(overlapping, axis=xaxis) > 0, rho, numpy.nan)
    
    return rho

def calc_overlap_adjacent(q, cf_min=0, cf_max=1):
    overlap = 0.0 * q.sum(axis=1)
    for k in range(1, q.crm_z.size):
        overlap[:, k] = calc_overlap(q[:, :, k], q[:, :, k - 1], xaxis=1)
    return overlap

def calc_rankcorr_adjacent(q):
    rankcorr = 0.0 * q.sum(axis=1)
    for k in range(1, q.crm_z.size):
        rankcorr[:, k] = calc_overlap(q[:, :, k], q[:, :, k - 1], xaxis=1)
    return rankcorr
```

```python
# TODO: this seems to be better, cleaner, and faster than my original implementation...replace
# the old modify_mmf code with this here?
def param_overlap(z, z0):
    overlap = numpy.zeros(z.shape)
    for k in range(1, z.shape[1]):
        overlap[:, k] = numpy.exp(-(z[:, k - 1] - z[:, k]) / z0)
    return overlap


def param_variance(qmean, a, b):
    qvar = a * qmean ** b
    return qvar


def regen_subcol(ds, overlap='maximum-random', condensate='homogeneous', padj=1, seed=-1,
                 variance='calculated',
                 alphac_z0=1.4e3, alphap_z0=1.4e3, rhoc_z0=1e3, rhop_z0=1e3):
    
    # need to load to reindex DataArrays
    ds.load()
    
    # get original dimension order
    original_dims = ds.dims
    
    # make sure grid goes from TOA to surface
    if ds.crm_z[0] > ds.crm_z[-1]:
        ds = ds.reindex(crm_z=ds.crm_z[::-1])
        flip = True
    else:
        flip = False
        
    # working variables
    qdict = {}
    for n in ('CRM_QC', 'CRM_QI', 'CRM_QPC', 'CRM_QPI'):
        if 'lon' in ds.dims:
            qdict[n] = ds[n].stack(pts=('time', 'lat', 'lon'), col=('crm_x', 'crm_y'))
        else:
            qdict[n] = ds[n].rename({'time': 'pts'}).stack(col=('crm_x', 'crm_y'))
        qdict[n] = qdict[n].transpose('pts', 'col', 'crm_z')
    
    # height coordinate for calculating overlap and rankcorr
    if 'lon' in ds.dims:
        z = ds['Z3'].stack(pts=('time', 'lat', 'lon')).transpose('pts', 'lev')
    else:
        z = ds['Z3'].rename({'time': 'pts'})
    z = z.reindex(lev=ds.crm_z).rename({'lev': 'crm_z'})
    
    # total condensate amounts
    qct = qdict['CRM_QC'] + qdict['CRM_QI']
    qpt = qdict['CRM_QPC'] + qdict['CRM_QPI']
    
    # cloud and precipitation fraction
    cf = (qct > 0).mean('col')
    pf = (qpt > 0).mean('col')
    
    # calculate gridbox means
    means = {}
    for qn in ('CRM_QC', 'CRM_QI'):
        means[qn] = qdict[qn].where(qct > 0).mean('col') 
    for qn in ('CRM_QPC', 'CRM_QPI'):
        means[qn] = qdict[qn].where(qpt > 0).mean('col')
    
    # calculate gridbox variance
    qvars = {}
    if condensate == 'homogeneous':
        for qn in means.keys():
            qvars[qn] = 0.0 * means[qn]
    elif condensate == 'calculated':
        for qn in ('CRM_QC', 'CRM_QI'):
            qvars[qn] = qdict[qn].where(qct > 0).var('col')
        for qn in ('CRM_QPC', 'CRM_QPI'):
            qvars[qn] = qdict[qn].where(qpt > 0).var('col')
    elif condensate == 'parameterized':
        qvars['CRM_QC'] = (0.44 * means['CRM_QC'] ** 0.71) ** 2
        qvars['CRM_QI'] = (0.90 * means['CRM_QI'] ** 1.02) ** 2
        qvars['CRM_QPC'] = (1.39 * means['CRM_QPC'] ** 0.95) ** 2
        qvars['CRM_QPI'] = (1.54 * means['CRM_QPI'] ** 0.99) ** 2
    else:
        raise NameError('Variance not recognized.')
        
    # calculate overlap parameters
    if overlap == 'maximum-random':
        alphac = -numpy.ones(cf.shape)
    elif overlap == 'maximum':
        alphac = numpy.ones(cf.shape)
    elif overlap == 'random':
        alphac = numpy.zeros(cf.shape)
    elif overlap == 'calculated':
        alphac = numpy.nan_to_num(calc_overlap_adjacent(qct))
    elif overlap == 'parameterized':
        alphac = param_overlap(z, alphac_z0)
    else:
        raise NameError('Overlap not recognized.')
    alphap = -numpy.ones(cf.shape)  # use prec_scops

    # calculate rank correlation
    if condensate == 'homogeneous':
        rhoc = -numpy.ones(cf.shape)
        rhop = -numpy.ones(pf.shape)
    elif condensate == 'calculated':
        rhoc = numpy.nan_to_num(calc_rankcorr_adjacent(qct))
        rhop = numpy.nan_to_num(calc_rankcorr_adjacent(qpt))
    elif condensate == 'parameterized':
        rhoc = param_overlap(z, rhoc_z0)
        rhop = param_overlap(z, rhop_z0)
    else:
        raise NameError('Condensate not recognized.')
    
    # call subcolumn generator
    qc_out, qi_out, qpc_out, qpi_out = subcol.gen_subcol(
        ds.crm_x.size, cf, pf, 
        means['CRM_QC'], means['CRM_QI'], means['CRM_QPC'], means['CRM_QPI'],
        qvars['CRM_QC'], qvars['CRM_QI'], qvars['CRM_QPC'], qvars['CRM_QPI'],
        alphac, alphap, rhoc, rhop, padj=padj, seed=seed
    )
    
    # copy regenerated condensate back to dict
    qdict['CRM_QC'].values = qc_out
    qdict['CRM_QI'].values = qi_out
    qdict['CRM_QPC'].values = qpc_out
    qdict['CRM_QPI'].values = qpi_out
    
    # unstack dimensions and copy back to dataset
    for n in qdict.keys():
        if 'pts' in ds.dims:
            ds[n] = qdict[n].unstack('pts').unstack('col')
        else:
            ds[n] = qdict[n].rename({'pts': 'time'}).unstack('col')
    
    # flip height grid back
    if flip:
        ds = ds.reindex(crm_z=ds.crm_z[::-1])
    
    # reorder dimensions so they match original
    ds = ds.transpose(*original_dims)
    
    return ds


def homogenize(ds):
    for n in ('CRM_QC', 'CRM_QI', 'CRM_QPC', 'CRM_QPI'):
        mean = ds[n].where(ds[n] > 0).mean('crm_x')
        axis = ds[n].dims.index('crm_x')
        ds[n][:] = numpy.where(ds[n] > 0, numpy.expand_dims(mean, axis), 0)
    return ds


def resample(ds):
    ds_out = ds.copy(deep=True)
    indices = numpy.random.rand.randint(0, high=ds.crm_x.size, ds.crm_x.size)
    for n in ('CRM_QC', 'CRM_QI', 'CRM_QPC', 'CRM_QPI'):
        ds_out[n] = ds[n].sel(crm_x=indices)
    return ds_out


def create_case(ds, case, seed=-1):
    # make a copy
    ds = ds.copy(deep=True)
    
    if case == 'crm':
        ds_out = ds
    elif case == 'crm-hom':
        ds_out = homogenize(ds)
    elif case == 'mro-hom':
        ds_out = regen_subcol(
            ds, overlap='maximum-random', condensate='homogeneous', padj=0, seed=seed
        )
    elif case == 'mro-hom-padj':
        ds_out = regen_subcol(
            ds, overlap='maximum-random', condensate='homogeneous', padj=1, seed=seed
        )
    elif case == 'gen-hom-calc':
        ds_out = regen_subcol(
            ds, overlap='calculated', condensate='homogeneous', padj=1, seed=seed
        )
    elif case == 'gen-var-calc':
        ds_out = regen_subcol(
            ds, overlap='calculated', condensate='calculated', padj=1, seed=seed,
        )
    elif case == 'gen-var-param':
        ds_out = regen_subcol(
            ds, overlap='parameterized', condensate='parameterized', padj=1, seed=seed,
        )
    else:
        raise NameError('Case not recongized!')
    return ds_out
```

```python
import subcol
cases = ('crm', 'crm-hom', 'mro-hom', 'gen-var-calc', 'gen-var-param')
figure, axes = matplotlib.pyplot.subplots(len(cases), 2, figsize=(8, 6))
with xarray.open_dataset(test_file) as ds:
    for i, case in enumerate(cases):
        ds_test = create_case(ds, case, seed=1)
        ds_test = ds_test.sel(lat=test_lat, lon=test_lon, method='nearest')
    
        qct = ds_test['CRM_QC'] + ds_test['CRM_QI']
        qpt = ds_test['CRM_QPC'] + ds_test['CRM_QPI']
        for j, q in enumerate((qct, qpt)):
            ax = figure.add_axes(axes[i, j])
            import matplotlib.colors
            pl = ax.pcolor(
                q.squeeze().transpose('crm_z', 'crm_x'), 
                cmap=matplotlib.pyplot.get_cmap('plasma'), 
                vmin=1e-5, vmax=1e1,
                norm=matplotlib.colors.LogNorm()
            )
            ax.text(0.99, 0.99, case, transform=ax.transAxes, ha='right', va='top')
            ax.set_xlim(0, ds.crm_x.size)
            ax.set_ylim(0, ds.crm_z.size)
            ax.set_xticklabels('')
            ax.set_yticklabels('')
figure.tight_layout()            
cb = figure.colorbar(pl, ax=axes.ravel().tolist(), fraction=0.02, label='Mixing ratio (g/kg)')
```
